```python
import torch
import torchvision
import torchvision.transforms as T
from PIL import Image
```

```python
img = torch.zeros(3, 256, 256)*255
transform = T.ToPILImage()
img = transform(img)
img
```

```python
img = torch.ones(3, 256, 256)
transform = T.ToPILImage()
img = transform(img)
img
```

```python
img = torch.rand(3, 256, 256)
transform = T.ToPILImage()
img = transform(img)
img
```

```python
aang = torchvision.io.read_image("/content/Two_copper_pieces.png")/255
print(aang.shape)
#aang
transform(aang)
```

```python
noise = torch.zeros(aang.shape)
distorted = aang+noise
transform(distorted)
#torch.equal(distorted, aang)
```

```python
noise = torch.rand(aang.shape)/9
distorted = aang+noise
transform(distorted)
```

```python
disses = [0.0, 0.1, 0.2, 0.5, 0.9, 1.0]
for diss in disses:
  noise = torch.rand(aang.shape)*diss
  distorted = aang+noise
  distorted = distorted
  print(str(diss))
  display(transform(distorted))
```

```python
disses = [0.0, 0.1, 0.2, 0.5, 0.9, 1.0]
for diss in disses:
  noise = torch.rand(aang.shape)*diss
  distorted = aang+noise
  print(str(diss))
  transform(distorted).save(f"/content/Haggaling_Aangs/aang_distorted_{str(diss)}.png")
```
